# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
import pandas as pd 
data = pd.read_csv('movie_bd_v5.csv')
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

data['profit'] = data.revenue - data.budget
data['release_date'] = pd.to_datetime(data['release_date'], format='%m/%d/%Y')
data['month'] = pd.DatetimeIndex(data['release_date']).month
data['original_title_len'] = data['original_title'].apply(lambda s: len(s))
data['genres'] = data['genres'].apply(lambda x:x.split('|'))
data['director'] = data['director'].apply(lambda x: x.split('|'))                                                    
data['cast'] = data['cast'].apply(lambda x: x.split('|'))
data['production_companies'] = data['production_companies'].apply(lambda x: x.split('|'))   

df_production = data.explode('production_companies')
df_production.groupby(['production_companies'])['original_title_len'].mean().sort_values()
```

```python
answers['24'] = 'Four By Two Productions'
```
